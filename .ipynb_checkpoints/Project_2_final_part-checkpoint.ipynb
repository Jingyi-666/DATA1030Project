```python
# import packages
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold

import matplotlib
from matplotlib import pylab as plt
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline

from sklearn.metrics import log_loss
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier

from sklearn.model_selection import ParameterGrid
from sklearn.model_selection import GridSearchCV
```

```python
# load dataset and clean data
df = pd.read_csv('data/heart_2022_with_nans.csv')

df = df.drop(columns=['HeightInMeters','WeightInKilograms','RemovedTeeth',\
                      'ChestScan','FluVaxLast12','PneumoVaxEver','TetanusLast10Tdap'])
df_no_null = df.dropna(axis=0)

# create feature matrix and target variable
X = df_no_null.loc[:, df_no_null.columns != 'HadDepressiveDisorder']

y = df_no_null['HadDepressiveDisorder']
y = y.replace({'Yes': 1, 'No': 0})
```

```python
print('X: ', X.shape)
print(X.columns)
print('y: ', y.shape)
```

*Remark: Small changes made in feature selection, so different number of columns from the midterm part.*

```python
# collect the various features
onehot_ftrs = ['State','Sex','PhysicalActivities','HadHeartAttack','HadAngina','HadStroke','HadAsthma','HadSkinCancer','HadCOPD',\
               'HadKidneyDisease','HadArthritis','HadDiabetes','DeafOrHardOfHearing','BlindOrVisionDifficulty','DifficultyConcentrating',\
               'DifficultyWalking','DifficultyDressingBathing','DifficultyErrands','RaceEthnicityCategory','AlcoholDrinkers','HIVTesting',\
               'HighRiskLastYear','CovidPos']
ordinal_ftrs = ['GeneralHealth','LastCheckupTime','SmokerStatus','ECigaretteUsage','AgeCategory']
ordinal_cats = [['NA','Poor','Fair','Good','Very good','Excellent'],\
                ['Within past year (anytime less than 12 months ago)','Within past 2 years (1 year but less than 2 years ago)',\
                 'Within past 5 years (2 years but less than 5 years ago)','5 or more years ago','NA'],
                ['NA','Never smoked','Former smoker','Current smoker - now smokes every day','Current smoker - now smokes some days'],\
                ['NA','Never used e-cigarettes in my entire life','Not at all (right now)','Use them some days','Use them every day'],\
                ['NA','Age 18 to 24','Age 25 to 29','Age 30 to 34','Age 35 to 39','Age 40 to 44','Age 45 to 49','Age 50 to 54',\
                 'Age 55 to 59','Age 60 to 64','Age 65 to 69','Age 70 to 74','Age 75 to 79','Age 80 or older']]
minmax_ftrs = ['PhysicalHealthDays','MentalHealthDays','SleepHours']
std_ftrs = ['BMI']

# collect all the encoders
preprocessor = ColumnTransformer(
    transformers=[
        ('ord', OrdinalEncoder(categories = ordinal_cats), ordinal_ftrs),
        ('onehot', OneHotEncoder(sparse_output=False,handle_unknown='ignore'), onehot_ftrs),
        ('minmax', MinMaxScaler(), minmax_ftrs),
        ('std', StandardScaler(), std_ftrs)])
prep = Pipeline(steps=[('preprocessor', preprocessor)])
final_scaler = StandardScaler()
```

```python
nr_states = 5

baseline_scores = []
test_scores = []
```

### ML algorithm: Logistic Regression

```python
# use GridSearchCV; ML algorithm: Logistic Regression
alpha = np.logspace(-2,2,8)
rho = np.linspace(0, 1, 8, dtype='f2')

param_grid = {
              'logisticregression__C': 1 / alpha, 
              'logisticregression__l1_ratio': rho, 
              'logisticregression__class_weight': [None, 'balanced']
              } 

X_tests_logreg = []
y_tests_logreg = []
results_logreg = []
best_models_logreg = []
final_models_logreg = []
test_scores_logreg = np.zeros(nr_states)
baseline_scores_logreg = np.zeros(nr_states)

for i in range(nr_states):
    print('random state =', 42*i)
    
    # first split to get a stratified sample
    X_sample, X_else, y_sample, y_else = train_test_split(X,y,stratify=y,test_size = 0.9,random_state=42*i)
    # second split to separate out the test set
    X_other, X_test, y_other, y_test = train_test_split(X_sample,y_sample,stratify=y_sample,test_size = 0.2,random_state=42*i)
    # stratified kfold splitter for other
    skf = StratifiedKFold(n_splits=4,shuffle=True,random_state=42*i)
    
    # save test sets
    X_tests_logreg.append(X_test)
    y_tests_logreg.append(y_test)
    
    # initialize the classifier
    clf = LogisticRegression(penalty='elasticnet',random_state = 42*i,max_iter=10000,solver='saga')
    # the pipeline
    pipe = make_pipeline(prep,final_scaler,clf)
    
    # use GridSearchCV
    grid = GridSearchCV(pipe,param_grid=param_grid,scoring='neg_log_loss',cv=skf,return_train_score=True,verbose=True)
    
    # fit the model on other
    grid.fit(X_other, y_other)
        
    # save results into a data frame
    results_logreg.append(pd.DataFrame(grid.cv_results_))
    
    # find best model
    best_models_logreg.append(grid.best_params_)
    print('best model parameters:',grid.best_params_)
    print('best validation score:',-grid.best_score_)
    
    # save the model
    final_models_logreg.append(grid)
    
    # calculate and save the test score
    y_test_pred = grid.predict(X_test)
    y_test_pred_prob = grid.predict_proba(X_test)
    test_scores_logreg[i] = log_loss(y_test,y_test_pred_prob)
    print('test score:',test_scores_logreg[i])
    
    # baseline score
    y_prob = np.mean(y_test)
    y_prob_baseline = np.full_like(y_test, y_prob)
    baseline_scores_logreg[i] = log_loss(y_test, y_prob_baseline)

import pickle
file = open('results/logreg.save', 'wb')
pickle.dump((results_logreg,best_models_logreg,final_models_logreg,X_tests_logreg,y_tests_logreg),file)
file.close()
```

```python
feature_names_logreg = grid.feature_names_in_
baseline_scores.append(baseline_scores_logreg)
test_scores.append(test_scores_logreg)
```

### ML algorithm: XGBoost

```python
# use GridSearchCV; ML algorithm: XGBoost
param_grid = {
              "reg_alpha": [0e0, 1e-2, 1e-1, 1e0, 1e1, 1e2],
              "reg_lambda": [0e0, 1e-2, 1e-1, 1e0, 1e1, 1e2],
              "max_depth": [1,3,10,30,100],
              } 

X_tests_xgb = []
X_test_preps_xgb = []
y_tests_xgb = []
results_xgb = []
best_models_xgb = []
final_models_xgb = []
test_scores_xgb = np.zeros(nr_states)
baseline_scores_xgb = np.zeros(nr_states)

for i in range(nr_states):
    print('random state =', 42*i)
    
    # first split to get a stratified sample
    X_sample, X_else, y_sample, y_else = train_test_split(X,y,stratify=y,test_size = 0.9,random_state=42*i)
    # second split to separate out the train
    X_train, X_other, y_train, y_other = train_test_split(X_sample,y_sample,stratify=y_sample,test_size = 0.4,random_state=42*i)
    # third split to separate out the train and validation set
    X_val, X_test, y_val, y_test = train_test_split(X_other,y_other,stratify=y_other,test_size = 0.5,random_state=42*i)
    
    # preprocess
    prep = Pipeline(steps=[('preprocessor', preprocessor), ('finalscaler', final_scaler)])
    X_train_prep = prep.fit_transform(X_train)
    X_val_prep = prep.transform(X_val)
    X_test_prep = prep.transform(X_test)
    
    # save test sets
    X_tests_xgb.append(X_test)
    X_test_preps_xgb.append(X_test_prep)
    y_tests_xgb.append(y_test)
    
    # initialize the classifier
    XGB = XGBClassifier(early_stopping_rounds=50,learning_rate=0.03,n_estimators=10000,seed=0,colsample_bytree=0.9,subsample=0.66)
    # use GridSearchCV
    grid = GridSearchCV(XGB,param_grid=param_grid,scoring='neg_log_loss',cv=2,return_train_score=True,verbose=True)
    
    # fit the model on other
    grid.fit(X_train_prep, y_train,eval_set=[(X_val_prep, y_val)],verbose=False)
        
    # save results into a data frame
    results_xgb.append(pd.DataFrame(grid.cv_results_))
    
    # find best model
    best_models_xgb.append(grid.best_params_)
    print('best model parameters:',grid.best_params_)
    print('best validation score:',-grid.best_score_)
    
    # save the model
    final_models_xgb.append(grid)
    
    # calculate and save the test score
    y_test_pred = grid.predict(X_test_prep)
    y_test_pred_prob = grid.predict_proba(X_test_prep)
    test_scores_xgb[i] = log_loss(y_test,y_test_pred_prob)
    print('test score:',test_scores_xgb[i])
    
    # baseline score
    y_prob = np.mean(y_test)
    y_prob_baseline = np.full_like(y_test, y_prob)
    baseline_scores_xgb[i] = log_loss(y_test, y_prob_baseline)

import pickle
file = open('results/xgb.save', 'wb')
pickle.dump((results_xgb,best_models_xgb,final_models_xgb,X_tests_xgb,X_test_preps_xgb,y_tests_xgb),file)
file.close()
```

```python
baseline_scores.append(baseline_scores_xgb)
test_scores.append(test_scores_xgb)
```

### Visualize test scores

```python
# compute mean & std of baseline & test scores
baseline_means = []
means = []
stds = []

for b in baseline_scores:
    baseline_means.append(np.mean(b))

for t in test_scores:
    means.append(np.mean(t))
    stds.append(np.std(t))

print(baseline_means)
print(means)
print(stds)
```
